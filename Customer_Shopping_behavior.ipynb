```python
# Now since there is not missing value we will continue with our processing read to use for sql
# by changing the column name to snake cassing like from Customer ID to customer_id. for better usuability
```

```python
df.columns = df.columns.str.lower() # changes the string to lower case
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns= {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create the new column age_group
```

```python
labels= ['Young Adult', 'Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months': 90
} 
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df['purchase_frequency_days'].head(10)
```

```python
# now checking the column discount applied and promo code used, becuase it is not necessarily always true that discount is given only after promo code

df[['discount_applied','promo_code_used']].head(10)
```

```python
# for check if both the column is exactly same
(df['discount_applied']==df['promo_code_used']).all()   # returns true so that we can remove promo code column 
```

```python
df = df.drop('promo_code_used',axis = 1)
```
